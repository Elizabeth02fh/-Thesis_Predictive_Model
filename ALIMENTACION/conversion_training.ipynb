```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/TESIS_PYTHON/ALIMENTACION/balanced_food_selection.csv')
```

```python
df.sample(3)
```

```python
#Alimentación: Reemplazando con valores ordinales
df['v_frutas'].replace((2,1,0),('5 porciones de verduras y 4 frutas','4 porciones de verduras y 3 frutas','3 porciones de verduras y 2 frutas'), inplace = True)
df['r_grasas'].replace((0,1,2,3),('Nada','Poco', 'Mucho', 'Demasiado'), inplace = True)
df['l_etiquetas'].replace((0,1,2,3,4),('Nunca','Casi nunca','A veces', 'Frecuentemente', 'Siempre'), inplace = True)
df['b_alcoholicas'].replace((4,3,2,1,0),('Nunca', 'Una vez al año', 'Menos de 12 veces al año', 'Una vez al mes', 'Cada fin de semana'), inplace = True)
df['agua'].replace((0,1,2),('Menos de 6 vasos', 'Entre 6 a 8 vasos', 'Más de 8 vasos'), inplace = True)
df['comer_deshoras'].replace((0,1,2,3,4),('Presentación de proyectos Finales', 'Exámenes Parciales','Tengo que trabajar y estudiar', 'Vivo lejos a la universidad','Falta de apoyo económico'), inplace = True)
#Target: Reemplazando con valores binarios
df['Diagnostico'].replace((0,1),('No Saludable', 'Saludable'),inplace = True)
```

```python
df.sample(3)
```

```python
df.to_csv('/content/drive/MyDrive/TESIS_PYTHON/ALIMENTACION/cate_balanced_food_selection.csv', index=False)
```
