```python
import pandas as pd
import numpy as np
```

```python
#conversion del test.csv a categoricos
df1 = pd.read_csv('/content/drive/MyDrive/TESIS_PYTHON/ALIMENTACION/food_test.csv')
```

```python
df1.sample(3)
```

```python
df1=df1.drop(['c_dulces','peso'], axis=1)
df1.sample(3)
```

```python
#Alimentación: Reemplazando con valores ordinales
df1['v_frutas'].replace((2,1,0),('5 porciones de verduras y 4 frutas','4 porciones de verduras y 3 frutas','3 porciones de verduras y 2 frutas'), inplace = True)
df1['r_grasas'].replace((0,1,2,3),('Nada','Poco', 'Mucho', 'Demasiado'), inplace = True)
df1['l_etiquetas'].replace((0,1,2,3,4),('Nunca','Casi nunca','A veces', 'Frecuentemente', 'Siempre'), inplace = True)
df1['b_alcoholicas'].replace((4,3,2,1,0),('Nunca', 'Una vez al año', 'Menos de 12 veces al año', 'Una vez al mes', 'Cada fin de semana'), inplace = True)
df1['agua'].replace((0,1,2),('Menos de 6 vasos', 'Entre 6 a 8 vasos', 'Más de 8 vasos'), inplace = True)
df1['comer_deshoras'].replace((0,1,2,3,4),('Presentación de proyectos Finales', 'Exámenes Parciales','Tengo que trabajar y estudiar', 'Vivo lejos a la universidad','Falta de apoyo económico'), inplace = True)
#Target: Reemplazando con valores binarios
df1['Diagnostico'].replace((0,1),('No Saludable', 'Saludable'),inplace = True)
```
