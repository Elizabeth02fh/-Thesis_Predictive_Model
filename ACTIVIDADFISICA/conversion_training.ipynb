```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/TESIS_PYTHON/ACTIVIDADFISICA/balanced_activity_selection.csv')
```

```python
df.sample(3)
```

```python
#Acitividad Física: Reemplazando con valores ordinales
df['a_moderadas'].replace((0,1,2),('Menos de 2 h y 30 min','2 h y 30 min a 5 h', 'Más de 5 h'), inplace = True)
df['a_intensas'].replace((0,1,2),('Menos de 1 h y 15 min', '1 h y 15 min a 2 h y 30 min','Más de 2 h y 30 min'), inplace = True)
df['f_muscular'].replace((0,1,2),('Menos de 2 días', 'Igual a 2 días', 'Más de 2 días'), inplace = True)
df['f_tendones'].replace((0,1,2,3),('Ninguno','Menos de 15 s', '15 a 30 s','Más 30 s'), inplace =  True)
#Target: Reemplazando con valores binarios
df['Diagnostico'].replace((0,1),('No Saludable', 'Saludable'),inplace = True)
```
